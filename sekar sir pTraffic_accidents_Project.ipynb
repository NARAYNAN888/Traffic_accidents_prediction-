```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,roc_curve,roc_auc_score,silhouette_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from statistics import mode
d=pd.read_csv('trafficaccident.csv')
d
df = d.drop([ 'Location_Easting_OSGR', 'Location_Northing_OSGR','Longitude','Latitude', 'Police_Force','Date','Local_Authority_(District)','Local_Authority_(Highway)','1st_Road_Class', '1st_Road_Number', 'Speed_limit','Junction_Detail', 'Junction_Control', '2nd_Road_Class','2nd_Road_Number', 'Pedestrian_Crossing-Human_Control','Pedestrian_Crossing-Physical_Facilities','Special_Conditions_at_Site', 'Carriageway_Hazards','Did_Police_Officer_Attend_Scene_of_Accident','Accident_Index','LSOA_of_Accident_Location','Urban_or_Rural_Area','Time'], axis=1)
df
tr=d['Urban_or_Rural_Area']



# x_train,x_test,y_train,y_test=train_test_split(df,tr,train_size=0.8)

# df
```

```python
df['Weather_Conditions']=df['Weather_Conditions'].fillna(mode(df['Weather_Conditions']))
df['Road_Surface_Conditions']=df['Road_Surface_Conditions'].fillna(mode(df['Road_Surface_Conditions']))
df.isna().sum()
```

```python
df['Light_Conditions']=LabelEncoder().fit_transform(df['Light_Conditions'])

df['Road_Type']=LabelEncoder().fit_transform(df['Road_Type'])
df

df['Weather_Conditions']=LabelEncoder().fit_transform(df['Weather_Conditions'])

df['Road_Surface_Conditions']=LabelEncoder().fit_transform(df['Road_Surface_Conditions'])

df

```

```python
from sklearn.cluster import KMeans,MiniBatchKMeans,DBSCAN,Birch

import numpy as np
Y=np.array((d['Urban_or_Rural_Area']))
Y=Y.reshape(-1,1)
X = df[['Accident_Severity','Number_of_Vehicles','Number_of_Casualties','Day_of_Week','Road_Type','Light_Conditions','Weather_Conditions','Road_Surface_Conditions','Year']].copy()
```

```python
kmeans = KMeans(n_clusters=2, random_state=100)
kmeans.fit(X)
pr=kmeans.predict(X)
ac=accuracy_score(Y,pr)
si=silhouette_score(Y,pr)
print(pr)
print(ac)
print(si)
```

```python
bb=Birch(threshold=0.5, branching_factor=50, n_clusters=2, compute_labels=True, copy=True)
bb.fit(X)
pr4=bb.fit_predict(X)
ac4=accuracy_score(Y,pr)
si4=silhouette_score(Y,pr)
print(pr4)
print(ac4)
print(si4)
```

```python
db=DBSCAN(eps=0.5, min_samples=2, metric='euclidean',leaf_size=10)
db.fit(X)
pr2=db.fit_predict(X,sample_weight=None)
ac2=accuracy_score(Y,pr)
si2=silhouette_score(Y,pr)
print(pr2)
print(ac2)
print(si2)
```

```python
mn=MiniBatchKMeans(n_clusters=2, max_iter=100, batch_size=1024, verbose=0, compute_labels=True, random_state=10, tol=0.0, max_no_improvement=10)
mn.fit(X)
pr3=mn.predict(X,sample_weight=None)
ac3=accuracy_score(Y,pr)
si3=silhouette_score(Y,pr)
print(pr3)
print(ac3)
print(si3)
```
